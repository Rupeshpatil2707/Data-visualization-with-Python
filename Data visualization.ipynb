# Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Bivariate data

### A.Line plot

```python
x=[1,2,3,4,5,6,7]
y=[1,4,9,16,25,36,49]
z=[1,8,27,64,125,216,343]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title('Line Plot',fontsize=10)
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.plot(x,y,color='#06841f',   #Use color names also
         linestyle='dashed',
         marker='o',
         markersize=10,
         markerfacecolor='#f1548e'
         
        )
plt.show()
```

```python
plt.title('Line Plot',fontsize=14)
plt.xlabel('X-Values')
plt.ylabel('Y-Values')
plt.plot(x,y,color='#06841f',   #Use color names also
         linestyle='solid',
         marker='^',
         markersize=10,
         markerfacecolor='#f1548e',
         label='IND'
        )

plt.plot(x,z,color='red',   #Use color names also
         linestyle='dotted',
         marker='o',
         markersize=10,
         markerfacecolor='blue',
        label='PAK')

plt.legend()
plt.show()
```

### B.Scatter Plot

```python
price=pd.read_csv('house_prices.csv')
price.head()
```

```python
price.rename(columns={'Area (ft.)':'Area'},inplace=True)
```

## Numerical Vs Categorical

```python
courses=['Data Science','Data Analytics','Tableau','Power BI','Mysql']
students=[70,57,63,81,45]
```

```python
plt.bar(courses,students)
plt.show()
```

```python
plt.figure(figsize=(7,6))  # Size of figure
plt.title('Bar Chart')
plt.bar(courses,students,
        color='green'
       ,alpha=.75 #Alpha means Transparency color
       
       )
plt.xticks(fontsize=10,rotation=45)   #Xticks rotation angle
plt.show()
```

```python
plt.figure(figsize=(7,6))  # Size of figure
plt.title('Bar Chart')
plt.barh(courses,students,
        color='#c191f9'
       ,alpha=.75 #Alpha means Transparency color
       )
plt.xticks(fontsize=10,rotation=45)   #Xticks rotation angle
plt.show()
```

### C.Pie Plot

```python
plt.pie(students,labels=courses)
plt.show()
```

```python
plt.title("Pie Chart")

plt.pie(students,labels=courses,
        autopct='%0.2f%%',  # Digit change by 2/1/3/4
        textprops={'fontsize':10,'color':'Blue'},
        explode=(0,0,0.1,0,0)    # Explode pulled out 
       )

plt.show()
```

# Catplot
## Bar Plot

```python
df_cars=pd.read_csv('bar_chart_data.csv')
df_cars
```

```python
pal=sns.color_palette('viridis')
sns.catplot(data=df_cars,x='Brand',y='Cars Listings',
            kind='bar',
           aspect=2/1,
           palette='winter'
           )
plt.show()
```

```python
sns.catplot(data=df_cars,x='Cars Listings',y='Brand',
            kind='bar',
           aspect=2/1,
           palette='viridis'
           )
plt.show()
```

## Boxplot

```python
sns.catplot(data=tips,x='day',y='total_bill',
           kind='box')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',
           kind='box',hue='time')     #Hue for give color based on types
plt.show()
```

### Strip Plot

```python
sns.catplot(data=tips,x='day',y='total_bill',
           kind='strip'    #hue='time'
           )     #Hue for give color based on types
plt.show()
```

### Swarm Plot

```python
sns.catplot(data=tips,x='day',y='total_bill',
           kind='swarm',hue='day'
           )     #Hue for give color based on types
plt.show()
```

```python
titanic=sns.load_dataset('titanic')
iris=sns.load_dataset('iris')
```

```python
sns.boxplot(data=iris,orient='h')
plt.show()
```

# Relplot
### Lineplot

```python
mpg=pd.read_csv('mpg.csv')
mpg.head()
```

```python
sns.relplot(data=mpg,x='model_year',y='mpg',
            kind='line',ci=False,   # CI=confidence intervel
            palette='rainbow',hue='origin',
            style='origin'
           )
plt.show()
```
